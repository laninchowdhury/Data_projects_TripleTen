# Model Building and Evaluation:

```python
def smape(y_true, y_pred):
    return 100 * np.mean(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))

def weighted_smape(y_true_rougher, y_pred_rougher, y_true_final, y_pred_final):
    alpha = 0.5
    smape_rougher = smape(y_true_rougher, y_pred_rougher)
    smape_final = smape(y_true_final, y_pred_final)
    return alpha * smape_rougher + (1 - alpha) * smape_final
```

```python
# Define Models
models ={
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor()
}
# Evaluate Cross_validation
scorer = make_scorer(smape, greater_is_better=False)

#Identify columns that are present in both training and test set
common_columns = x_train.columns.intersection(x_test.columns)
x_train_filtered = x_train[common_columns]
x_test_filtered = x_test[common_columns]

print("Evaluating models for fougher recovery:")
for model_name, model in models.items():
    scores = cross_val_score(model, x_train_filtered, y_train_rougher, cv=5, scoring=scorer)
    print(f"{model_name}(Rougher)sMAPE:{-np.mean(scores):.2f} ")
    
print("\nEvaluating models for final recovery:")    
for model_name, model in models.items():
    scores = cross_val_score(model, x_train_filtered, y_train_final, cv=5, scoring=scorer)
    print(f"{model_name}(Final) sMAPE:{-np.mean(scores):.2f} ")
```
