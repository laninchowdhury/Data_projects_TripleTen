```python
# import some necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.utils import resample
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

### Load and Prepare the Data

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('https://practicum-content.s3.us-west-1.amazonaws.com/datasets/Churn.csv')

# Inspect the first few rows and data types
print(data.head())
print(data.info())
```

### Drop Unnecessary Columns

```python
# Drop unnecessary columns
columns_to_drop = ['RowNumber', 'CustomerId', 'Surname']
data = data.drop(columns=columns_to_drop)
```

```python

# Handle missing values in 'Tenure'
data['Tenure'].fillna(data['Tenure'].median(), inplace=True)

# One-Hot Encode Categorical Features
categorical_columns = ['Geography', 'Gender']
data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)

# Verify columns after encoding
print(data.columns)

# Verify there are no missing values
print(data.isnull().sum())
```

### Define Features and Target

```python
# Define features and target
features = data.drop('Exited', axis=1)
target = data['Exited']

# Verify shapes
print("Features shape:", features.shape)
print("Target shape:", target.shape)
```

### Split the data

```python
from sklearn.model_selection import train_test_split

# Split data into training (60%), validation (20%), and test (20%) sets
features_train, features_temp, target_train, target_temp = train_test_split(features, target, test_size=0.4, random_state=12345, stratify=target)
features_valid, features_test, target_valid, target_test = train_test_split(features_temp, target_temp, test_size=0.5, random_state=12345, stratify=target_temp)

# Verify the number of samples in each set
print(f"Training set shape: {features_train.shape} {target_train.shape}")
print(f"Validation set shape: {features_valid.shape} {target_valid.shape}")
print(f"Test set shape: {features_test.shape} {target_test.shape}")
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize Numerical Features
scaler = StandardScaler()
features_train_scaled = scaler.fit_transform(features_train)
features_valid_scaled = scaler.transform(features_valid)
features_test_scaled = scaler.transform(features_test)

# Verify transformed shapes
print("Transformed training set shape:", features_train_scaled.shape)
print("Transformed validation set shape:", features_valid_scaled.shape)
print("Transformed test set shape:", features_test_scaled.shape)
```

### Train the Model Without Handling Class Imbalance

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, roc_auc_score

# Train Logistic Regression model
model_lr = LogisticRegression(random_state=12345, max_iter=1000)
model_lr.fit(features_train_scaled, target_train)

# Evaluate on Validation Set
predictions_valid_lr = model_lr.predict(features_valid_scaled)
f1_lr = f1_score(target_valid, predictions_valid_lr)
roc_auc_lr = roc_auc_score(target_valid, model_lr.predict_proba(features_valid_scaled)[:, 1])

print("Initial Logistic Regression F1 Score (Validation):", f1_lr)
print("Initial Logistic Regression ROC-AUC Score (Validation):", roc_auc_lr)
```

### Handle Class Imbalance with Oversampling

```python
from sklearn.utils import resample

# Separate majority and minority classes
features_train_majority = features_train[target_train == 0]
features_train_minority = features_train[target_train == 1]

# Reset indices before resampling
features_train_majority.reset_index(drop=True, inplace=True)
features_train_minority.reset_index(drop=True, inplace=True)

# Random oversample minority class
features_train_minority_oversampled, target_train_minority_oversampled = resample(
    features_train_minority, target_train[target_train == 1], replace=True, n_samples=len(features_train_majority), random_state=12345)

# Combine majority and oversampled minority
features_train_oversampled = pd.concat([features_train_majority, features_train_minority_oversampled], axis=0)
target_train_oversampled = pd.concat([target_train[target_train == 0], target_train_minority_oversampled], axis=0)

# Standardize Numerical Features
features_train_oversampled_scaled = scaler.fit_transform(features_train_oversampled)

# Train Logistic Regression model on oversampled data
model_lr_oversampled = LogisticRegression(random_state=12345, max_iter=1000)
model_lr_oversampled.fit(features_train_oversampled_scaled, target_train_oversampled)

# Evaluate on Validation Set
predictions_valid_lr_oversampled = model_lr_oversampled.predict(features_valid_scaled)
f1_lr_oversampled = f1_score(target_valid, predictions_valid_lr_oversampled)
roc_auc_lr_oversampled = roc_auc_score(target_valid, model_lr_oversampled.predict_proba(features_valid_scaled)[:, 1])

print("Logistic Regression with Oversampling F1 Score (Validation):", f1_lr_oversampled)
print("Logistic Regression with Oversampling ROC-AUC Score (Validation):", roc_auc_lr_oversampled)
```

### Handle Class Imbalance with Undersampling

```python
# Random undersample majority class
features_train_majority_undersampled, target_train_majority_undersampled = resample(
    features_train_majority, target_train[target_train == 0], replace=False, n_samples=len(features_train_minority), random_state=12345)

# Combine undersampled majority and minority
features_train_undersampled = pd.concat([features_train_majority_undersampled, features_train_minority], axis=0)
target_train_undersampled = pd.concat([target_train_majority_undersampled, target_train[target_train == 1]], axis=0)

# Standardize Numerical Features
features_train_undersampled_scaled = scaler.fit_transform(features_train_undersampled)

# Train Random Forest Classifier on undersampled data
from sklearn.ensemble import RandomForestClassifier

model_rf_undersampled = RandomForestClassifier(random_state=12345, n_estimators=100, max_depth=10)
model_rf_undersampled.fit(features_train_undersampled_scaled, target_train_undersampled)

# Evaluate on Validation Set
predictions_valid_rf_undersampled = model_rf_undersampled.predict(features_valid_scaled)
f1_rf_undersampled = f1_score(target_valid, predictions_valid_rf_undersampled)
roc_auc_rf_undersampled = roc_auc_score(target_valid, model_rf_undersampled.predict_proba(features_valid_scaled)[:, 1])

print("Random Forest Classifier with Undersampling F1 Score (Validation):", f1_rf_undersampled)
print("Random Forest Classifier with Undersampling ROC-AUC Score (Validation):", roc_auc_rf_undersampled)
```

<h2>   Reviewer's comment 3 </h2>
    
Let me clarify the idea of conducting the final test on **the best model**.

The test subset (usually called hold-out subset) that is saved for the final test is used for an objective evaluation of the model's quality after training and hyperparameter tuning. The test subset contains data that was not used during either the model's training or hyperparameter tuning. This allows us to assess how well the model can perform on new, previously unseen data. When tuning hyperparameters on the validation subset, it is possible to accidentally overfit the model to this set. By keeping the test subset isolated from this process, we can ensure that the model has not been overfitted.

Moreover, in real-world tasks, we often evaluate our models on real-time data that could not have been known during the model's training process. By reserving the test set for a final evaluation, we simulate this situation. This is why it is crucial not to use the test subset during the training process.

Additionally, it is important to select the best model based on metrics obtained from the validation subset. This is because we often deal with hgue datasets and heavy models, making it computationally expensive to run every model we trained on the hold-out test subset.

    
    
So please train `XGBClassifier` in the previous section on the training subset, like you did with the Forest models, then compare the result with the Forest F1 scores and choose the best model (**1 model**) that yielded the highest F1 on validation.
